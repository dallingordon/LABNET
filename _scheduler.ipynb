```python
import numpy as np
import matplotlib.pyplot as plt
import inspect
```

```python
min_lr = 5e-8
max_lr = 5e-5
dist_min = np.log10(min_lr)
dist_max = np.log10(max_lr)
dist_min,dist_max
```

```python
decay_epochs = 800
final_max = 5e-6
dist_final_max = np.log10(final_max)
unit_decay = dist_max - dist_final_max
unit_decay,dist_final_max
```

```python
epoch_decay = [(i,max(decay_epochs-i,0)) for i in range(1000)]
epoch_decay 
```

```python
#i[0] is epoch, i[1] is decayreductions (in colab trainer.)
data = [10**np.random.uniform(dist_min, dist_max) for i in epoch_decay]
data = [10**np.random.uniform(dist_min, dist_max-(1-(i[1]/decay_epochs))*unit_decay) for i in epoch_decay]
```

```python
fig, axs = plt.subplots(2, 1, figsize=(18, 10))
x = range(len(data))
# Plotting the line graph on the first subplot
axs[0].plot(x, data, linestyle='-', color='b')
axs[0].set_title("Line Graph of Data")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")

# Plotting the histogram on the second subplot
axs[1].hist(data, bins=20, color='c', edgecolor='black')
axs[1].set_title("Histogram of Data")
axs[1].set_xlabel("Value")
axs[1].set_ylabel("Frequency")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the graphs
plt.show()
```
